```python
from codeassist import WizardCoder
```

```python
baselines = [
    "Using python langchain library, summarize a given article using stuff strategy.",
    "Using python langchain library, start a conversation with openai chat api using model gpt-3.5-turbo and ask it convert a given english text into french.",
    "Using python langchain library, split a text into chunks of 300 words each with overlap of 50 words.",
    "Using python langchain library, create a sql to query how many unique product ids are there in the products table."
]
```

```python
model = WizardCoder(model_name_or_path="WizardLM/WizardCoder-3B-V1.0", fp16=True, bf16=False)
```

```python
for baseline in baselines:
  result = model.generate(baseline)
  print(baseline)
  print("---")
  print(result[0])
  print("---\n\n")
```
